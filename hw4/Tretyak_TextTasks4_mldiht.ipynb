$
\rho(\omega, \sigma) = \frac{1}{(2\pi\sigma)^{n/2}}e^{-\frac{||\omega||^2}{2\sigma}}, ||\omega||^2 = \sum_{k=1}^n {w_k}^2
$

Логарифмируя, получим регуляризатор по $L_2$ норме
$
-ln(\rho(\omega, \sigma)) = \frac{1}{2\sigma}||w||^2 + const(\omega)
$

Пусть $\omega $ имеет n-мерное распределение лапласа:

$
\rho(\omega, C) = \frac{1}{(2C)^{n}}e^{-\frac{||\omega||_1}{2\sigma}}, ||\omega||_1 = \sum_{k=1}^n |w_k|
$

Логарифмируя, получим регуляризатор по $L_1$ норме
$
-ln(\rho(\omega, C)) = \frac{1}{C}||w||_1 + const(\omega)
$

3) Сделаем веса такими, чтобы 

$
\min_{i=1=..l}= y_i(<w,x_i> - w_0) = 1
$

Ширина полосы равна:
$
<(x_+ - x_-), \frac{w}{||w||}> = \frac{2}{||w||}
$
, где $x_+$ и $x_-$ - точки на границе полосы. Тогда, в случае неразделимой выборки, получаем:

$$ \left\{
\begin{aligned}
<w,w>->min\\
y_i(<w,x_i> - w_0) \ge 1, i = 1..l\\
\end{aligned}
\right. $$

$$ \left\{
\begin{aligned}
\frac{1}{2}<w,w> + C\sum_{i=1}^l \xi_i->min\\
y_i(<w,x_i> - w_0) \ge 1 - \xi_i, i = 1..l\\
\xi_i \ge 0, i = 1..l\\
\end{aligned}
\right. $$

Заметим, что:
$
M_i = y_i(<w,x_i> - w_0) 
$

$
\xi_i \ge 0
$

$
\xi_i \ge 1 - M_i
$

$
\sum_{i=1}^l \xi_i->min
$
Тогда, 

$
\xi_i = (1 - M_i)_+
$
Тогда, 

$
Q(w,w_0) = \sum_{i=1}^l (1-M_i(w,w_0))_+ + \frac{1}{2C}<w,w> ->min_{w,w_0}
$
